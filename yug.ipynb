```python
import matplotlib.pyplot as plt
```

```python
plt.plot(r_shape.index,r_shape["java"])
```

```python
plt.figure(figsize=(16,8))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.plot(r_shape.index,r_shape["java"],label="java")
```

```python
plt.figure(figsize=(16,10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("year",fontsize=14)
plt.ylabel("no of posts",fontsize=14)
plt.ylim(0,35000)
plt.plot(r_shape.index,r_shape["java"],label="java")
```

```python
plt.figure(figsize=(16,10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("year",fontsize=14)
plt.ylabel("no of posts",fontsize=14)
plt.ylim(0,35000)
plt.plot(r_shape.index,r_shape["java"],label="java")
plt.plot(r_shape.index,r_shape["python"],label="python")
```

```python
plt.figure(figsize=(16,10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("year",fontsize=14)
plt.ylabel("no of posts",fontsize=14)
plt.ylim(0,35000)
for colum in r_shape.columns:
  plt.plot(r_shape.index,r_shape[colum],label=colum,linewidth=3)


```

```python
plt.figure(figsize=(16,10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("year",fontsize=14)
plt.ylabel("no of posts",fontsize=14)
plt.ylim(0,35000)
for colum in r_shape.columns:
  plt.plot(r_shape.index,r_shape[colum],label=colum,linewidth=3)
plt.legend(fontsize=16)
plt.show()

```
